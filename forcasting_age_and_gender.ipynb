```python
# с помощью этой функции будем разбивать на слова колонку all_words
def tokenize(text):
    tokens = text.split(';')
    res = []
    for i in tokens:
        if i.strip():
            for j in i.split(' '):
                if j.strip().find('&referrer=')>-1:j = j.strip()[10:]
                if j.strip().find('&referrer=http:')>-1: j= j.strip()[15:]  
                if j.strip().find('${referer_url}')>-1: j= j.strip()[14:]
                if j.strip(): res.append(j.strip().lower())  
    return res
```
